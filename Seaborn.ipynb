```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
import pandas as pd

data = pd.read_csv("C:\\Users\\devis\\OneDrive\\Desktop\\python data analysis\\seaborn-data-master\\tips.csv")

print(data.head())
```

```python
#Plot a scatter plot between total_bill and tip
sns.relplot(x='total_bill',y='tip',data=data)
```

```python
#the hue parameter
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

```python
#the style parameter
sns.relplot(x='total_bill',y='tip',hue='smoker',style="sex",data=data)
```

```python
#the size parameter
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(15,200),data=data)
```

```python
#Categorical Scatterplot(Stripplot and Swarmplot)
# x-aix-- Categorical data
#y-axis-- Numerical data
```

```python
plt.style.use('fivethirtyeight')
```

```python
#Strip plot
sns.catplot(x='day',y='tip',kind='strip',data=data)
```

```python
#Swarm plot
sns.catplot(x='day',y='tip',kind='swarm',hue='sex',data=data)
```

```python
#Boxplot
```

```python
#univariate
sns.boxplot(data['tip'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='box',hue='sex',data=data)
```

```python
#Violin Plots
```

```python
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',split=True,data=data)
```

```python
#Barplot and Countplot
```

```python
#barplot
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',estimator=np.std,kind='bar',hue='sex',data=data)
```

```python
#countplot
sns.catplot(x='sex',kind='count',hue='smoker',data=data)
```

```python
#Distplots
data=pd.read_csv(r"C:\Users\devis\OneDrive\Desktop\python data analysis\seaborn-data-master\titanic.csv")
```
